```python
import spacy
import json
import pylab
from IPython.core.display import display, HTML
import nltk
from tika import parser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from spacytextblob.spacytextblob import SpacyTextBlob

%matplotlib inline
pylab.rcParams['figure.figsize'] = (10., 8.)
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
```

```python
import PyPDF2
from tika import parser

def extract_text_with_pypdf2(pdf_path):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
        return text

def extract_text_with_tika(pdf_path):
    parsed = parser.from_file(pdf_path)
    return parsed['content']
```

```python
pdf_path1 = 'QM2InitialComplaint.pdf'
pdf_path2 = 'QM2MotionToDismiss.pdf'

text1_pypdf2 = extract_text_with_pypdf2(pdf_path1)
text2_pypdf2 = extract_text_with_pypdf2(pdf_path2)

text1_tika = extract_text_with_tika(pdf_path1)
text2_tika = extract_text_with_tika(pdf_path2)
```

```python
%%capture

import scattertext as st

df = pd.DataFrame({'text': [text1_tika, text2_tika], 'category': ['PDF1', 'PDF2']})

corpus = st.CorpusFromPandas(
    df,
    category_col='category',
    text_col='text',
    nlp=nlp  # Assuming you have loaded your spacy model as 'nlp'
).build()

```

```python
html = st.produce_scattertext_explorer(
    corpus,
    category='PDF1',  # Category for the x-axis
    category_name='PDF1',
    not_category_name='PDF2',  # Category for the y-axis
    width_in_pixels=1000,
    metadata=df['category']  # Optional metadata
)

display(HTML(html))
```
